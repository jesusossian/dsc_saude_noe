```python
# Instalar bibliotecas
#!pip install pandas openpyxl
import pandas as pd
import os
```

```python
file_path = "../data/base em analise.xlsx"
```

```python
# 3Ô∏è‚É£ Ler Excel
df = pd.read_excel(file_path, engine="openpyxl", header=0)
```

```python
df
```

```python
# 4Ô∏è‚É£ Normalizar nomes das colunas
df.columns = df.columns.str.lower().str.strip()
df.columns
```

```python
# 5Ô∏è‚É£ Procurar colunas de volume
possible_cols_mes = [c for c in df.columns if 'volume (m3) por m√™s' in c]
possible_cols_ano = [c for c in df.columns if 'volume (m3) por ano' in c]

print(possible_cols_mes)
print(possible_cols_ano)

if not possible_cols_mes:
    raise ValueError("Erro: N√£o existe coluna 'Volume (m3) por m√™s' no ficheiro.")

if not possible_cols_ano:
    raise ValueError("Erro: N√£o existe coluna 'Volume (m3) por ano' no ficheiro.")

col_mes = possible_cols_mes[0]
col_ano = possible_cols_ano[0]
```

```python
# 6Ô∏è‚É£ Calcular volumes totais
total_mes = df[col_mes].sum()
total_ano = df[col_ano].sum()

print(total_mes)
print(total_ano)
```

```python
# 7Ô∏è‚É£ Listas de medicamentos
arvs = ["Tenofovir", "Lamivudina", "Efavirenz"]
antimalaricos = ["Artemether-Lumefantrina", "Cloroquina", "Quinina"]
antibioticos = ["Amoxicilina", "Ciprofloxacina", "Azitromicina"]

todos_medicamentos = arvs + antimalaricos + antibioticos
```

```python
# 8Ô∏è‚É£ Distribuir volumes igualmente pelos 9 medicamentos
vol_por_mes = total_mes / len(todos_medicamentos)
vol_por_ano = total_ano / len(todos_medicamentos)
```

```python
# 9Ô∏è‚É£ Construir tabela final
result = pd.DataFrame({
    "Medicamento": todos_medicamentos,
    "Volume (m¬≥) por m√™s": [vol_por_mes] * len(todos_medicamentos),
    "Volume (m¬≥) por ano": [vol_por_ano] * len(todos_medicamentos)
})
```

```python
# üîü Mostrar resultados
print("\n=== QUANTIDADES CALCULADAS COM BASE NO VOLUME TOTAL ===")
print(result)
```
